# Question 1: Load a CSV Dataset into a Pandas DataFrame

In this step, the CSV dataset is loaded into a Pandas DataFrame using the Pandas library for further data analysis and preprocessing.

```python
# Loading CSV dataset

df = pd.read_csv("tshirts.csv")

print("Dataset loaded successfully!")
```

# Question 3: Handle Missing Values (identify, fill/drop)

In this step, missing values are identified and handled using suitable techniques.

```python
# Checking missing values

print(df.isnull().sum())
```

```python
# Selecting numeric columns

numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

print("Numeric Columns:")
print(numeric_columns)
```

```python
# Filling numeric missing values with mean

for col in numeric_columns:
    
    df[col] = df[col].fillna(df[col].mean())

print("Numeric missing values handled successfully!")
```

```python
# Selecting categorical columns

categorical_columns = df.select_dtypes(include=['object']).columns

print("Categorical Columns:")
print(categorical_columns)
```

```python
# Filling categorical missing values with mode

for col in categorical_columns:
    
    df[col] = df[col].fillna(df[col].mode()[0])

print("Categorical missing values handled successfully!")
```

```python
# Checking missing values again

print(df.isnull().sum())
```

# Question 4: Perform Basic Operations (filter rows, select columns)

In this step, basic data operations such as column selection and row filtering are performed.

```python
# Selecting first two columns

selected_columns = df.iloc[:, :2]

print(selected_columns.head())
```

```python
# Filtering rows using numeric column

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

filtered_data = df[df[numeric_cols[0]] > df[numeric_cols[0]].mean()]

print(filtered_data.head())
```

# Question 5: Remove Duplicates

In this step, duplicate rows are identified and removed from the dataset.

```python
# Checking duplicate rows

print("Duplicate Rows:", df.duplicated().sum())
```

```python
# Removing duplicate rows

df = df.drop_duplicates()

print("Duplicate rows removed successfully!")
```

```python
print(df.columns)
```

# Question 6: Create a Derived Column (total_amount = price × quantity)

In this step, a new derived column named total_amount is created using the price and quantity columns.

```python
# Creating price and quantity columns

df['price'] = df['initial_price'] - df['discount']

df['quantity'] = df['ratings_count']
```

```python
# Creating total_amount column

df['total_amount'] = df['price'] * df['quantity']

print("New column created successfully!")

print(df[['price', 'quantity', 'total_amount']].head())
```

# Question 7: Save Cleaned Dataset

In this step, the cleaned dataset is saved as a new CSV file for future use.

```python
# Saving cleaned dataset

df.to_csv("cleaned_tshirts.csv", index=False)

print("Cleaned dataset saved successfully!")
```
